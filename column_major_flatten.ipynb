```python
import torch
import numpy
import pandas as pd
```

```python
x_min=0
x_max=2
t_min=0
t_max=2
total_points_x=4
total_points_t=4
```

```python
x = torch.linspace(x_min, x_max, 4).view(-1,1)
t = torch.linspace(t_min, t_max, 4).view(-1,1)
```

```python
X, T = torch.meshgrid(x.squeeze(1),t.squeeze(1))
print(X)
print("---------------------------------------")
print(T)
```

```python
x_test = torch.hstack((X.flatten()[:,None],T.flatten()[:,None])) #row major
pd.DataFrame(x_test, columns=["X","T"])
```

```python
x_test_T = torch.hstack((X.T.flatten()[:,None],T.T.flatten()[:,None])) #Column major which simulates Time Marching
pd.DataFrame(x_test_T, columns=["X","T"])
```

```python
x_test_Tt = torch.hstack((X.transpose(1,0).flatten()[:,None],T.transpose(1,0).flatten()[:,None])) #Column major which simulates Time Marching
pd.DataFrame(x_test_Tt, columns=["X","T"])
```
